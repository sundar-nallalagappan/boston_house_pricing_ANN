### Import the required libraries

```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
print('Tensorflow version is {}'.format(tf.__version__))
```

### Load the boston housing dataset

```python
(features, actual_price), (test_features, test_prices) = tf.keras.datasets.boston_housing.load_data(test_split=0.2)

print('Numer of data points         :', features.shape[0])
print('Numer of independent features:', features.shape[1])
print('Numer of dependent features  :', actual_price.shape[0])
```

### Building the graph in Keras - Instantiating the sequential model

```python
# Instantiate the Schema - Sequentail model - output from layer is input to another layer
model = tf.keras.models.Sequential()

# Normalize Input data
model.add(tf.keras.layers.BatchNormalization(input_shape=(13,)))

# Stack one hidden layer
model.add(tf.keras.layers.Dense(units=7, activation='relu'))

# Output layer
model.add(tf.keras.layers.Dense(units=1, activation='linear'))
```

### Compile the model defined above

```python
model.compile(optimizer='sgd', loss='mse')
```

### Summary of the model

```python
model.summary()
```

### Train the model

```python
train_model = model.fit(features, actual_price,  epochs=750, verbose=2)
```

#### Make predictions for test data points

```python
y_pred = model.predict(test_features)
```

#### Evaluate the model

```python
metrics.r2_score(test_prices, y_pred.ravel())
```
